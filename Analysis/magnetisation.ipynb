# Magnetisation

### Magnetisation with initial Mn3+ configuration for P- and N- substitutions

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('P_conc.csv')
df1 = pd.read_csv('N_conc.csv')

fig, ax = plt.subplots(1,2, figsize=(13, 7))
ax[0].scatter(x = df['Conc'], y = df['mag_per_atom'], edgecolors='r')
ax[0].set_xlabel("P_Conc",fontsize=20)
ax[0].set_ylabel("magnetisation per atom",fontsize=20)

ax[1].scatter(x = df1['Conc'], y = df1['mag_per_atom'], edgecolors='k')
ax[1].set_xlabel("N_conc",fontsize=20)
ax[1].set_ylabel("magnetisation per atom",fontsize=20)

plt.show()
```

### Magnetisation with initial Mn4+ configuration for P- and Mn3+ for N- substitutions

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('P_conc_d3.csv')
df1 = pd.read_csv('N_conc.csv')

fig, ax = plt.subplots(1,2, figsize=(13, 7))
ax[0].scatter(x = df['Conc'], y = df['mag_per_atom'], edgecolors='r')
ax[0].set_xlabel("P_Conc",fontsize=20)
ax[0].set_ylabel("magnetisation per atom",fontsize=20)

ax[1].scatter(x = df1['Conc'], y = df1['mag_per_atom'], edgecolors='k')
ax[1].set_xlabel("N_conc",fontsize=20)
ax[1].set_ylabel("magnetisation per atom",fontsize=20)

plt.show()
```
